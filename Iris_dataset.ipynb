## 데이터 정의
---

* sklearn 으로 부터 iris 데이터셋을 불러옵니다.
* x_data와 y_data를 정의합니다.

```python
! ls
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('iris_train.csv')
data
```

```python
data_np = data.to_numpy()
print(data_np)
```

```python
x_data= data_np[:,2:].astype('float')
y_data= data_np[:,1]
print(x_data.shape, y_data.shape)
```

```python
#'setosa' 0,
#'versicolor' 1,
#'virginica' 2

y_data_label =[]
for y in y_data:
  if y == 'setosa':
    y_data_label.append(0)
  elif y == 'versicolor':
    y_data_label.append(1)
  else:
    y_data_label.append(2)
print(y_data_label)
```

## 데이터 시각화
---

* 데이터를 matplotlib.pyplot 라이브러리를 이용항 시각화 합니다.


```python
plt.plot(x_data, y_data, marker = 'o', linestyle='')
plt.legend(['sepal length', 'petal length', 'sepal width','petal width'])
plt.show()
```

## 데이터 전처리
---

* y 데이터 값을 원-핫 벡터로 코딩합니다.
* 학습데이터와 테스트데이터를 분리합니다. 
  * sklearn.model_selection에서 제공하는 train_test_split 를 사용합니다.


### **One-hot 코딩**


```python
from tensorflow.keras.utils import to_categorical
Y_onehot = to_categorical(y_data_label, num_classes=3)
print(Y_onehot)
```

### **train-test split**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, Y_onehot, test_size=0.2, random_state=42)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

## 4. 모델학습
---

* softmax regression을 사용하여 훈련하고, 테스트셋을 이용해 성능을 평가합니다.
* decision tree를 사용하여 훈련하고, 테스트셋을 이용해 성능을 평가합니다.
* 이 데이터셋에 적합한 세번째 모델을 찾아보고 훈련해 봅니다.

### **Softmax regression**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
model.add(Dense(3,input_dim=4,activation='softmax'))
model.summary()

model.compile(optimizer=optimizers.SGD(learning_rate=0.01), loss='categorical_crossentropy', metrics=['acc'])
print('# Fit model on training data')
history = model.fit(x_train, y_train, epochs=300)
```

```python
plt.plot(history.history['acc'])
plt.legend(['acc'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.legend(['loss'])
plt.show()
```

```python
print('\n# Evaluate on test data')
results = model.evaluate(x_test, y_test)
print('test loss, test acc:', results)
```

### **Decision Tree**

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```
